```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', 200)

df = pd.read_csv('../data/raw/application_train.csv')

print("Shape:", df.shape)
df.head()
```

```python
print(df['TARGET'].value_counts())
print(df['TARGET'].value_counts(normalize=True) * 100)

sns.countplot(x='TARGET', data=df)
plt.title('Target Distribution (0 = Repaid, 1 = Default)')
plt.show()
```

```python
print("Data types:\n", df.dtypes.value_counts())
missing = df.isnull().sum().sort_values(ascending=False)
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
missing_df[missing_df['Missing Count'] > 0].head(20)
```

```python
cat_cols = df.select_dtypes(include='object').columns
print(f"Number of categorical columns: {len(cat_cols)}")
print(cat_cols.tolist())

for col in cat_cols:
    print(f"{col}: {df[col].nunique()} unique values")
    
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score
import lightgbm as lgb

for col in cat_cols:
    df[col] = df[col].astype('category')

X = df.drop(columns=['TARGET', 'SK_ID_CURR'])
y = df['TARGET']

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

params = {
    'objective': 'binary',
    'metric': 'auc',
    'is_unbalance': True,
    'learning_rate': 0.05,
    'num_leaves': 31,
    'verbose': -1
}
```

```python
import mlflow
import mlflow.lightgbm

mlflow.set_experiment("loan-default-prediction")

with mlflow.start_run(run_name="baseline_lightgbm"):

    mlflow.log_params(params)
    mlflow.log_param("n_splits", 5)
    mlflow.log_param("num_features", X.shape[1])

    oof_preds = np.zeros(len(X))
    auc_scores = []

    for fold, (train_idx, val_idx) in enumerate(skf.split(X, y)):
        X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
        y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

        train_data = lgb.Dataset(X_train, label=y_train, categorical_feature=list(cat_cols))
        val_data = lgb.Dataset(X_val, label=y_val, categorical_feature=list(cat_cols), reference=train_data)

        model = lgb.train(
            params,
            train_data,
            valid_sets=[val_data],
            num_boost_round=1000,
            callbacks=[lgb.early_stopping(50), lgb.log_evaluation(0)]
        )

        oof_preds[val_idx] = model.predict(X_val, num_iteration=model.best_iteration)
        fold_auc = roc_auc_score(y_val, oof_preds[val_idx])
        auc_scores.append(fold_auc)

        mlflow.log_metric(f"fold_{fold+1}_auc", fold_auc)
        print(f"Fold {fold+1} AUC: {fold_auc:.4f}")

    mean_auc = np.mean(auc_scores)
    std_auc = np.std(auc_scores)
    overall_auc = roc_auc_score(y, oof_preds)

    mlflow.log_metric("mean_cv_auc", mean_auc)
    mlflow.log_metric("std_cv_auc", std_auc)
    mlflow.log_metric("overall_oof_auc", overall_auc)

    mlflow.lightgbm.log_model(model, "model")

    print(f"\nMean CV AUC: {mean_auc:.4f} (+/- {std_auc:.4f})")
    print(f"Overall OOF AUC: {overall_auc:.4f}")
```

```python
installments = pd.read_csv('../data/raw/installments_payments.csv')
print("Shape:", installments.shape)
print("Unique clients:", installments['SK_ID_CURR'].nunique())
installments.head()
```

```python
# Days late (positive = paid late, negative/zero = paid on time or early)
installments['DAYS_LATE'] = installments['DAYS_ENTRY_PAYMENT'] - installments['DAYS_INSTALMENT']

# Payment shortfall (positive = underpaid)
installments['PAYMENT_DIFF'] = installments['AMT_INSTALMENT'] - installments['AMT_PAYMENT']

inst_agg = installments.groupby('SK_ID_CURR').agg({
    'DAYS_LATE': ['mean', 'max', 'sum'],
    'PAYMENT_DIFF': ['mean', 'sum', 'max'],
    'AMT_INSTALMENT': ['mean', 'sum'],
    'AMT_PAYMENT': ['mean', 'sum'],
})

inst_agg.columns = ['INSTAL_' + '_'.join(col).upper() for col in inst_agg.columns]
inst_agg.reset_index(inplace=True)

print("Installments aggregated shape:", inst_agg.shape)
inst_agg.head()
```

```python
pos_cash = pd.read_csv('../data/raw/POS_CASH_balance.csv')
print("Shape:", pos_cash.shape)
print("Unique clients:", pos_cash['SK_ID_CURR'].nunique())

pos_agg = pos_cash.groupby('SK_ID_CURR').agg({
    'SK_DPD': ['mean', 'max'],           # Days past due
    'SK_DPD_DEF': ['mean', 'max'],       # Days past due (with tolerance)
    'CNT_INSTALMENT_FUTURE': ['mean', 'min'],
})

pos_agg.columns = ['POS_' + '_'.join(col).upper() for col in pos_agg.columns]
pos_agg.reset_index(inplace=True)

print("POS aggregated shape:", pos_agg.shape)
pos_agg.head()
```

```python
credit_card = pd.read_csv('../data/raw/credit_card_balance.csv')
print("Shape:", credit_card.shape)
print("Unique clients:", credit_card['SK_ID_CURR'].nunique())

cc_agg = credit_card.groupby('SK_ID_CURR').agg({
    'AMT_BALANCE': ['mean', 'max'],
    'AMT_CREDIT_LIMIT_ACTUAL': ['mean', 'max'],
    'SK_DPD': ['mean', 'max'],
    'CNT_DRAWINGS_CURRENT': ['mean', 'sum'],
})

cc_agg.columns = ['CC_' + '_'.join(col).upper() for col in cc_agg.columns]
cc_agg.reset_index(inplace=True)

print("Credit card aggregated shape:", cc_agg.shape)
cc_agg.head()
```

```python
df = df.merge(inst_agg, on='SK_ID_CURR', how='left')
df = df.merge(pos_agg, on='SK_ID_CURR', how='left')
df = df.merge(cc_agg, on='SK_ID_CURR', how='left')

print("Shape after all merges:", df.shape)

# Quick high-signal ratio features
df['RATIO_CREDIT_INCOME'] = df['AMT_CREDIT'] / df['AMT_INCOME_TOTAL']
df['RATIO_ANNUITY_INCOME'] = df['AMT_ANNUITY'] / df['AMT_INCOME_TOTAL']
df['RATIO_CREDIT_ANNUITY'] = df['AMT_CREDIT'] / df['AMT_ANNUITY']
df['RATIO_EMPLOYED_BIRTH'] = df['DAYS_EMPLOYED'] / df['DAYS_BIRTH']

print("Final shape:", df.shape)
```

```python
X = df.drop(columns=['TARGET', 'SK_ID_CURR'])
y = df['TARGET']

for col in cat_cols:
    X[col] = X[col].astype('category')

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

with mlflow.start_run(run_name="full_feature_set"):

    mlflow.log_params(params)
    mlflow.log_param("n_splits", 5)
    mlflow.log_param("num_features", X.shape[1])
    mlflow.log_param("feature_set", "bureau + prev_app + installments + pos + credit_card + ratios")

    oof_preds = np.zeros(len(X))
    auc_scores = []

    for fold, (train_idx, val_idx) in enumerate(skf.split(X, y)):
        X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
        y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

        train_data = lgb.Dataset(X_train, label=y_train, categorical_feature=list(cat_cols))
        val_data = lgb.Dataset(X_val, label=y_val, categorical_feature=list(cat_cols), reference=train_data)

        model = lgb.train(
            params,
            train_data,
            valid_sets=[val_data],
            num_boost_round=1000,
            callbacks=[lgb.early_stopping(50), lgb.log_evaluation(0)]
        )

        oof_preds[val_idx] = model.predict(X_val, num_iteration=model.best_iteration)
        fold_auc = roc_auc_score(y_val, oof_preds[val_idx])
        auc_scores.append(fold_auc)

        mlflow.log_metric(f"fold_{fold+1}_auc", fold_auc)
        print(f"Fold {fold+1} AUC: {fold_auc:.4f}")

    mean_auc = np.mean(auc_scores)
    std_auc = np.std(auc_scores)
    overall_auc = roc_auc_score(y, oof_preds)

    mlflow.log_metric("mean_cv_auc", mean_auc)
    mlflow.log_metric("std_cv_auc", std_auc)
    mlflow.log_metric("overall_oof_auc", overall_auc)

    mlflow.lightgbm.log_model(model, "model")

    print(f"\nMean CV AUC: {mean_auc:.4f} (+/- {std_auc:.4f})")
    print(f"Overall OOF AUC: {overall_auc:.4f}")
```

```python
%pip install optuna

import optuna
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score
import lightgbm as lgb
import numpy as np

def objective(trial):
    params = {
        'objective': 'binary',
        'metric': 'auc',
        'is_unbalance': True,
        'verbose': -1,
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.1, log=True),
        'num_leaves': trial.suggest_int('num_leaves', 20, 150),
        'max_depth': trial.suggest_int('max_depth', 3, 12),
        'min_child_samples': trial.suggest_int('min_child_samples', 10, 100),
        'feature_fraction': trial.suggest_float('feature_fraction', 0.5, 1.0),
        'bagging_fraction': trial.suggest_float('bagging_fraction', 0.5, 1.0),
        'bagging_freq': trial.suggest_int('bagging_freq', 1, 7),
        'reg_alpha': trial.suggest_float('reg_alpha', 0.0, 5.0),
        'reg_lambda': trial.suggest_float('reg_lambda', 0.0, 5.0),
    }

    skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)  # 3 folds for speed during search
    auc_scores = []

    for train_idx, val_idx in skf.split(X, y):
        X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
        y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

        train_data = lgb.Dataset(X_train, label=y_train, categorical_feature=list(cat_cols))
        val_data = lgb.Dataset(X_val, label=y_val, categorical_feature=list(cat_cols), reference=train_data)

        model = lgb.train(
            params,
            train_data,
            valid_sets=[val_data],
            num_boost_round=1000,
            callbacks=[lgb.early_stopping(50), lgb.log_evaluation(0)]
        )

        preds = model.predict(X_val, num_iteration=model.best_iteration)
        auc_scores.append(roc_auc_score(y_val, preds))

    return np.mean(auc_scores)

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=30)  # 30 trials — adjust based on time available

print("Best AUC:", study.best_value)
print("Best params:", study.best_params)
```

```python
final_params = {
    'objective': 'binary',
    'metric': 'auc',
    'is_unbalance': True,
    'verbose': -1,
    'learning_rate': 0.04484675330692807,
    'num_leaves': 98,
    'max_depth': 3,
    'min_child_samples': 38,
    'feature_fraction': 0.8784228709717604,
    'bagging_fraction': 0.7317520943641929,
    'bagging_freq': 1,
    'reg_alpha': 2.5864604706352545,
    'reg_lambda': 4.49455105794037,
}

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

with mlflow.start_run(run_name="tuned_final_model"):

    mlflow.log_params(final_params)
    mlflow.log_param("n_splits", 5)
    mlflow.log_param("num_features", X.shape[1])
    mlflow.log_param("feature_set", "full_feature_set + optuna_tuned")

    oof_preds = np.zeros(len(X))
    auc_scores = []
    models = []

    for fold, (train_idx, val_idx) in enumerate(skf.split(X, y)):
        X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
        y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

        train_data = lgb.Dataset(X_train, label=y_train, categorical_feature=list(cat_cols))
        val_data = lgb.Dataset(X_val, label=y_val, categorical_feature=list(cat_cols), reference=train_data)

        model = lgb.train(
            final_params,
            train_data,
            valid_sets=[val_data],
            num_boost_round=2000,
            callbacks=[lgb.early_stopping(50), lgb.log_evaluation(0)]
        )

        oof_preds[val_idx] = model.predict(X_val, num_iteration=model.best_iteration)
        fold_auc = roc_auc_score(y_val, oof_preds[val_idx])
        auc_scores.append(fold_auc)
        models.append(model)

        mlflow.log_metric(f"fold_{fold+1}_auc", fold_auc)
        print(f"Fold {fold+1} AUC: {fold_auc:.4f}")

    mean_auc = np.mean(auc_scores)
    std_auc = np.std(auc_scores)
    overall_auc = roc_auc_score(y, oof_preds)

    mlflow.log_metric("mean_cv_auc", mean_auc)
    mlflow.log_metric("std_cv_auc", std_auc)
    mlflow.log_metric("overall_oof_auc", overall_auc)

    mlflow.lightgbm.log_model(models[-1], "model")

    print(f"\nMean CV AUC: {mean_auc:.4f} (+/- {std_auc:.4f})")
    print(f"Overall OOF AUC: {overall_auc:.4f}")
```

```python
# Retrain final model on full dataset (no held-out validation split)
final_train_data = lgb.Dataset(X, label=y, categorical_feature=list(cat_cols))

final_model = lgb.train(
    final_params,
    final_train_data,
    num_boost_round=int(np.mean([m.best_iteration for m in models]))  # use avg best_iteration from CV folds
)

print("Final model trained on full dataset")
print("Num trees:", final_model.num_trees())
```
